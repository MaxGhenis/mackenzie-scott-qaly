## New findings from the regression pass

1.  **SEVERITY: MINOR**  
    **LOCATION:** Introduction, rendered p. 1, citation shown as “Gamboa 2026”; `paper/references.bib`, `ap2026givingusa`.  
    **Problem:** The AP/Chronicle page names Rasheeda Childress, not Glenn Gamboa. Its page title is “Donors gave U.S. charities \$617 billion in 2025, according to the new Giving USA report” (the HTML title is “Philanthropic donations hit \$617B in 2025, says Giving USA 2026”), not the title in the BibTeX entry. The page does support \$6.65B against \$19.2B, so the quantitative claim is sound but the reference is misidentified.  
    **Fix:** Change the author to Rasheeda Childress and use the publisher’s displayed title (plus the June 23, 2026 publication date).

2.  **SEVERITY: MINOR**  
    **LOCATION:** `paper/references.bib`, `givewell2026impact`, `robinhood2025`, `constellation2025`, and `cega2023div`; rendered references pp. 13–14.  
    **Problem:** Several web references added in the revision use an access/reporting year as though it were the publication year. GiveWell’s page says published November 2021 and last updated November 2025, not 2026; Robin Hood’s structured metadata says published 2023 and modified 2024, not 2025; Constellation’s *2025 Investment Report* page says published February 27, 2026; and the CEGA page is dated March 18, 2022, not 2023. The claims these pages support are otherwise accurately characterized.  
    **Fix:** Use each page’s publication or last-update year as appropriate and put July 2026 only in an access-date field/note. For CEGA, also add the named author shown by the page if the chosen citation style permits it.

3.  **SEVERITY: MINOR**  
    **LOCATION:** Model, rendered p. 3, “replicated child-targeted cash quasi-experiments (Aizer et al. 2016) anchor the cash bucket.”  
    **Problem:** Aizer et al. is one quasi-experimental study of the Mothers’ Pension program. It supports a child-targeted cash anchor, but by itself does not support the plural “replicated … quasi-experiments” characterization; the additional studies discussed in `data/parameters.yaml` are not cited here or included in the paper bibliography.  
    **Fix:** Either write “a child-targeted cash quasi-experiment” or cite the additional quasi-experimental replications underlying the plural claim.
