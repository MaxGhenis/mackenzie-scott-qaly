# 

The six requested journal records were checked against DOI/publisher metadata. `cesarini2016wealth`, `miller2021medicaid`, `sommers2017medicaid`, `bailey2015war`, and `aizer2016cash` have the correct named authors, year, journal, volume/issue, page span, and DOI; `finkelstein2012oregon` has the correct eight named authors and all other requested fields but omits the corporate coauthor, the Oregon Health Study Group. The manuscript nevertheless needs major citation revision: the megagiving statistic is attached to a temporally incapable and bibliographically misidentified CNBC article, Sommers (2017) is inaccurately grouped with an ACA mortality study despite the exact misattribution warning in `SOURCES.md`, and the GiveWell cash citation does not establish the manuscript’s portfolio-scale multiple or diminishing-returns claims. The CEP evidence and HHS valuation are directionally usable after narrower attribution and corrected bibliographic details, while two journal entries are orphaned and several quantitative conventions or literature claims lack in-text support.

1.  **SEVERITY: MAJOR**  
    **LOCATION:** Abstract, p. 1, “roughly a third of American megagiving in 2025”; Introduction, p. 1, “by the Indiana University Lilly Family School of Philanthropy’s accounting, about a third of every megagift in America that year”; `references.bib`, `cnbc2025scott`.  
    **Problem:** The underlying statistic is about dollars, not the number of gifts: \$6.65 billion of Scott’s contributions represented about one-third of the \$19.2 billion *value* of 2025 megagifts. The current wording “a third of every megagift” is scope-inaccurate. More seriously, the cited December 2025 CNBC story predates Giving USA 2026 and contains neither the \$19.2 billion denominator nor the one-third comparison; it discusses Scott’s \$7.17 billion and Giving USA’s *2024 total-giving* figure. Its BibTeX is also misattributed: the [CNBC publisher page](https://www.cnbc.com/2025/12/13/mackenzie-scott-revealed-her-total-charitable-donations-for-2025.html) names Tom Huddleston Jr., not Lorie Konish, and its title is “MacKenzie Scott announced another \$7.1 billion in 2025 charitable donations—she’s now given away \$26.3 billion since 2019.” The one-third statistic appears in contemporaneous 2026 reporting based on Giving USA/Lilly data, including [Fortune](https://fortune.com/2026/06/25/mackenzie-scott-largest-megadonor-2025-7-billion-donations-giving-usa-iu-report/) and the [Associated Press/Chronicle of Philanthropy](https://apnews.com/article/giving-usa-report-philanthropy-2025-8363b76bc8cf854f6865c31129e8a4b1); the [Giving USA release](https://givingusa.org/giving-usa-charitable-giving-rose-to-617-20-billion-in-2025-surpassing-the-600-billion-mark-for-the-first-time/) verifies the distinct \$617.20 billion total-giving figure and the report’s formal title. This is the same class of source/claim mismatch that the project previously corrected.  
    **Concrete fix:** Replace the sentence with, for example, “Scott’s \$6.65 billion in qualifying contributions represented about one-third of the \$19.2 billion value of U.S. megagifts in 2025,” and cite a source that actually states those figures (or give the exact Giving USA 2026 page/table). Remove CNBC from this claim. If CNBC remains for the \$7.17 billion annual total, correct its author and title. Preserve the distinction between megagift value and total charitable giving.

2.  **SEVERITY: MAJOR**  
    **LOCATION:** Model, p. 3, “the ACA Medicaid mortality quasi-experiment (Miller et al. 2021; Sommers 2017).”  
    **Problem:** [Miller, Johnson, and Wherry (2021)](https://doi.org/10.1093/qje/qjab004) studies mortality after Affordable Care Act Medicaid expansions. [Sommers (2017)](https://doi.org/10.1162/ajhe_a_00080) is a cost-benefit reanalysis of the earlier Arizona, Maine, and New York expansions; it supplies the \$327,000–\$867,000 cost-per-life-saved arithmetic documented in `SOURCES.md`, not a second ACA mortality design. Grouping both citations under “the ACA … experiment” recurs in substance the attribution problem that `SOURCES.md` expressly warns against, even though it does not repeat the old NEJM citation verbatim.  
    **Concrete fix:** Separate the roles: cite Miller et al. (2021) for ACA-expansion mortality identification, and cite Sommers (2017) specifically for the cost-benefit reanalysis of pre-ACA state expansions. If the underlying pre-ACA mortality estimate itself is discussed, cite Sommers, Baicker, and Epstein (2012) for that estimate while retaining Sommers (2017) for the cost arithmetic.

3.  **SEVERITY: MAJOR**  
    **LOCATION:** Section 4.2, p. 7, “funding beyond identified room drives toward cash-transfer cost-effectiveness” and “the 20–40× range implied by GiveWell’s cash-transfer benchmark (GiveWell 2024).”  
    **Problem:** [GiveWell’s 2024 post](https://blog.givewell.org/2024/11/12/re-evaluating-the-impact-of-unconditional-cash-transfers/) says that its revised estimate of GiveDirectly’s flagship cash program is 3–4 times its *previous estimate* and that GiveWell’s current top programs are at least twice as cost-effective as the revised cash estimate. It does not report a 20–40× comparison with Scott’s portfolio, does not convert cash-transfer welfare into QALYs, and does not say that tens of billions of additional funding converge to cash-transfer cost-effectiveness. Thus the citation is accurate for the fact of a cash re-evaluation but not for the characterization or numerical range attached to it here.  
    **Concrete fix:** Label 20–40× as an author/model-derived scenario and show the complete like-for-like arithmetic, including how GiveWell’s units are translated into health-only QALYs; otherwise remove the range. Cite the 2024 post only for its actual 3–4× re-evaluation. Do not describe cash as an empirically established portfolio-scale “floor” without a source and a unit-comparability argument.

4.  **SEVERITY: MAJOR**  
    **LOCATION:** Section 4.2, p. 7, “frontier-priced opportunities are scarce, GiveWell directs hundreds of millions — not tens of billions — per year at these prices,” and the closing reference to “diminishing returns accounts in the cited literature.”  
    **Problem:** These are externally checkable claims but no cited source in the manuscript establishes GiveWell’s annual directed amount, the amount available at the quoted frontier prices, or a portfolio-scale diminishing-returns curve. `givewell2024cash` does none of those things. Calling the material “cited literature” overstates the citation coverage.  
    **Concrete fix:** Add separate citations to GiveWell’s annual metrics/grantmaking totals and its room-for-more-funding or funding-bar methodology, quote the relevant year and amount, and characterize the portfolio-scale compression explicitly as a scenario unless a source quantifies it. If no source supplies a 20–40× curve, delete that precision.

5.  **SEVERITY: MAJOR**  
    **LOCATION:** Model, p. 3, “Health effects … are converted to discounted QALYs under a common set of conventions (about 65 remaining life-years at baseline utility 0.87, 3% annual discounting).”  
    **Problem:** The values are uncited in the manuscript and are not a common conversion across the model. `data/parameters.yaml` uses roughly 26 remaining years and a 0.60/0.78/0.92 utility distribution for adult premature deaths, while 65 years and 0.87 are fixed child-death assumptions used for the global-health frontier; a third pair calibrates HHS VQALY discount-rate rescaling. The sentence therefore both overgeneralizes and conceals which conversion drives which result. GiveWell reports cost per life saved, not the 65-year or 0.87 inputs.  
    **Concrete fix:** Replace “common set” with separate adult and child conversion descriptions. Cite the adult life-expectancy and utility anchors already named in `SOURCES.md`, add an authoritative life-table source and health-utility source for the child values, and cite the normative source for the 3% rate. Identify 65 and 0.87 as author calibration assumptions if that is what they remain.

6.  **SEVERITY: MINOR**  
    **LOCATION:** Model, p. 3, CEP capacity-effects sentence; `references.bib`, `cep2023gifts`.  
    **Problem:** The report title is not exact, and the entry suppresses the named authors. The publisher’s title page gives Ellie Buteau, Elisha Smith Arrillaga, and Christina Im as authors and the title/subtitle as *Emerging Impacts: The Effects of MacKenzie Scott’s Large, Unrestricted Gifts: Results from Year Two of a Three-Year Study*, not “(Year 2).” The [CEP report](https://cep.org/wp-content/uploads/2023/11/BigGiftsStudy_Report_Y2_FNL.pdf) does document recipient leaders’ reports of stronger operational capacity, staffing, financial stability, and expanded programs, but it is a 41%-response survey plus interviews, not a causal estimate of marginal delivery or a measured realization multiplier above one.  
    **Concrete fix:** Correct the title and authors in BibTeX. Revise the prose to “self-reported capacity changes among surveyed Scott recipients,” and state that using those reports to permit a realization factor above one is an author judgment rather than an effect size estimated by CEP.

7.  **SEVERITY: MINOR**  
    **LOCATION:** Introduction, p. 2, and Results, p. 4, benefit-cost ratio language; `references.bib`, `aspe2026values`.  
    **Problem:** The [ASPE landing page and 2026 PDF](https://aspe.hhs.gov/reports/standard-ria-values) identify Aaron Kearsley as author and the title as *HHS Standard Values for Regulatory Analysis, 2026*, not *Standard Values for Regulatory Impact Analysis*. The source does support the valuation input: Table 3 gives a 3%-discount VQALY range of \$339,000/\$726,000/\$1,105,000 for 2026 impacts in constant 2025 dollars. It does not itself report the manuscript’s 4.9 benefit-cost ratio, and “the U.S. government’s value” is broader than the source’s status as HHS default recommendations for HHS regulatory impact analyses.  
    **Concrete fix:** Correct the author/title and cite Table 3. Say “using HHS’s 2026 central regulatory-analysis VQALY at a 3% discount rate” (and disclose the manuscript’s CPI conversion); label 4.9 and its interval as model outputs, not ASPE results.
